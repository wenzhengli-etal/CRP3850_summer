````python
plt.style.use('seaborn-v0_8-whitegrid')

# Group by year and month to get the average price
df_avg_price = df_combine.groupby(['TransYear', 'TransMonth'])['UnitPrice'].count().unstack()

# Plotting the heatmap
plt.figure(figsize=(18, 5))
sns.heatmap(df_avg_price, annot = True, fmt = ".0f", cmap='YlOrBr')
plt.title('Average Monthly Housing Prices (2012-2016)')
plt.xlabel('Month')
plt.ylabel('Year')
````

````python
annual_avg = df_combine.groupby('Year')[['Olympic Center', 'The Great Wall']].mean().reset_index()
annual_avg = annual_avg[(annual_avg['Year'] >= 2014) & (annual_avg['Year'] <= 2018)]

plt.figure(figsize=(14, 5))
plt.plot(annual_avg['Year'], annual_avg['Olympic Center'], marker='x', label='Olympic Center')
plt.plot(annual_avg['Year'], annual_avg['The Great Wall'], marker='x', label='The Great Wall')

plt.title('Trend of PM25: 2014 to 2018')
plt.xlabel('Year')
plt.ylabel('PM25')
plt.xticks(annual_avg['Year'])
plt.show()
````
